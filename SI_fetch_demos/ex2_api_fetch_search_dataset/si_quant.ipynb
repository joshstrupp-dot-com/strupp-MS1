```python
# for each row in physicalDescription, identify the first number that is followed by ' mm' and add it to a new column called "length", and turn that number into an integer.
df['length'] = df['physicalDescriptions'].str.extract(r'(\d+) mm', expand=False)
df['length'] = df['length'].astype(float)
print(df.tail(20))
```

```python
# for physicalDescription rows that have missing values for length, if the first number found in the physicalDescription is followed by ' cm', multiply that number by 10 and add it to the length column.
# and if the first number found in the physicalDescription is followed by ' m', multiply that number by 1000 and add it to the length column.
df['length'] = df['length'].fillna(0)
df['length'] = df['length'] + df['physicalDescriptions'].str.extract(r'(\d+) cm', expand=False).fillna(0).astype(float) * 10
df['length'] = df['length'] + df['physicalDescriptions'].str.extract(r'(\d+) m', expand=False).fillna(0).astype(float) * 1000
```

```python
# for rows that have weight = 0, look in physicalDescription for numbers that are followed by ' kg' and multiply that number by 1000 and add it to the weight column.
df['weight'] = df['weight'] + df['physicalDescriptions'].str.extract(r'(\d+) kg', expand=False).fillna(0).astype(float) * 1000
```

# Attempt 1 of confidence interval for rest of population   

```python
from scipy import stats;
import numpy as np

# Sample size
n = len(df)

# Population size
N = 500000

# Sample mean and standard deviation
sample_mean = df['weight'].mean()
sample_std = df['weight'].std(ddof=1)

# Standard error of the mean
sem = sample_std / np.sqrt(n)

# Confidence level

confidence_level = 0.95

# Degrees of freedom
df_degrees = n - 1

# t-score for the confidence level
t_score = stats.t.ppf((1 + confidence_level) / 2, df_degrees)

# Margin of error
margin_of_error = t_score * sem

# Confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Confidence Interval: {confidence_interval}")
```

This means that based on your sample, you can be 95% confident that the true average weight of the entire population lies between 292.98 and 1929.55 (units are whatever weight is measured in, like grams or kilograms).
